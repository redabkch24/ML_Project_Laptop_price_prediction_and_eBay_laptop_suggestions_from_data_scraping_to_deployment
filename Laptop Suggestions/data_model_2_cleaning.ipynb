```python
import pandas as pd
```

Importing raw data

```python
df = pd.read_csv('data_model_2.csv')
```

```python
df.columns
```

```python
df = df.drop('rating', axis=1)
```

```python
df.columns = ['title', 'price', 'links','images']
```

Removing nan value

```python
for col in df.columns:
  df = df[df[col]!='not found']
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

Keeping just price on $

```python
df = df[df['price'].str.startswith('US $', na=False)]
```

Cleaning and Converting price column

```python
for c in ['US','$',',','/ea']:
  df.price = df.price.str.replace(c,'')
```

```python
df.price = pd.to_numeric(df.price)
```

Saving dataset

```python
df.to_csv('data_model_2_clean.csv')
```
