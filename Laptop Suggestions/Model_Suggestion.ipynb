```python
import pandas as pd
```

Importing dataset

```python
df = pd.read_csv('data_model_2_clean.csv')
```

Model creation

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import NearestNeighbors
import numpy as np

prix = df[['price']].values

nn = NearestNeighbors(n_neighbors=5, metric='euclidean')
nn.fit(prix)
```

Testing

```python
prix_saisi = 200

distances, indices = nn.kneighbors([[prix_saisi]])
for i in indices[0]:
    print(df.iloc[i])
```

Model Pickler

```python
import joblib
joblib.dump(nn,filename='model_nn.joblib')
```
