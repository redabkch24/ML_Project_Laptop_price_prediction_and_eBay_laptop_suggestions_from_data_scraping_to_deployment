```python
import pandas as pd
```

Importing cleaning dataset

```python
df = pd.read_csv('dataset_20_12_no_encoder.csv')
```

```python
data = df[['brand','storage','storage_hdd','storage_ssd','screen_size','ram','cpu_speed','cpu_brand','os','gpu','etat','current_price']]
```

```python
data.head()
```

Variables encoder using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le_brand = LabelEncoder()
le_storage_type = LabelEncoder()
le_cpu_brand = LabelEncoder()
le_etat = LabelEncoder()
le_os = LabelEncoder()
le_gpu = LabelEncoder()

data.brand = le_brand.fit_transform(data[['brand']])
data.etat = le_etat.fit_transform(data[['etat']])
data.cpu_brand = le_cpu_brand.fit_transform(data[['cpu_brand']])
data.os = le_os.fit_transform(data[['os']])
data.gpu = le_gpu.fit_transform(data[['gpu']])
```

```python
data.head()
```

```python
data.corr()
```

Removing columns that have a low correaltion with price (current_price == price == target)

```python
X = data.drop(columns=["current_price",'brand','cpu_brand'])
y = data["current_price"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=5,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

Linear Regression



```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
yt = model.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
yt = model.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
model = RandomForestRegressor()
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
yt = model.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
model_gb = GradientBoostingRegressor()
cv_scores = cross_val_score(model_gb, X, y, cv=5, scoring='neg_mean_absolute_error')
model_gb.fit(X_train,y_train)
y_pred = model_gb.predict(X_test)
yt = model_gb.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```

AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
model = AdaBoostRegressor()
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
yt = model.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```

XGBRegressor

```python
from xgboost import XGBRegressor
model_xg = XGBRegressor()
model_xg.fit(X_train,y_train)
y_pred = model_xg.predict(X_test)
yt = model_xg.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred))
print('mae : ', mean_absolute_error(y_test,y_pred))
print('r2 : ', r2_score(y_test,y_pred))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt))
print('mae : ', mean_absolute_error(y_train,yt))
print('r2 : ', r2_score(y_train,yt))
```



Tuning the model using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':range(10,300,10)}
gs = GridSearchCV(model_gb,parameters,scoring='neg_mean_absolute_error',cv=5)
gs.fit(X_train,y_train)
gs.best_estimator_
```

```python
y_pred_gs = gs.best_estimator_.predict(X_test)
yt_gs = gs.best_estimator_.predict(X_train)
print('-----------------Test---------------')
print('mse : ', mean_squared_error(y_test,y_pred_gs))
print('mae : ', mean_absolute_error(y_test,y_pred_gs))
print('r2 : ', r2_score(y_test,y_pred_gs))
print('---------------Training------------')
print('mse : ', mean_squared_error(y_train,yt_gs))
print('mae : ', mean_absolute_error(y_train,yt_gs))
print('r2 : ', r2_score(y_train,yt_gs))
```
