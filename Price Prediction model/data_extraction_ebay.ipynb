```python
import pandas as pd
import re
```

Importing row data

```python
df = pd.read_csv('data_laptop_ebay_2.csv')
```

```python
df = df.drop_duplicates()
```

Removing nan values (in scraping a use 'not found' to replace nan values)

```python
for col in df.columns:
  df = df[df[col] != 'not found']
```

```python
df.title = df.title.str.lower()
df['info'] = df['info'].str.lower()
df.etat = df.etat.str.lower()
```

```python
df['info'].unique()
```

Brand Extract

```python
def extract_brand(title):
    pattern = r"(lenovo|hp|dell|samsung|acer|msi|toshiba|apple|huawei|asus|alienware|fujitsu|panasonic|toughpad)"
    match = re.search(pattern, title, re.IGNORECASE)
    return match.group(0) if match else None
```

```python
df['brand'] = df.title.apply(extract_brand)
```

cpu speed extract



```python
def extract_processor_speed(text):
    match = re.search(r"Processor Speed\s*([\d.]+)\s*GHz", text, re.IGNORECASE)
    return match.group(1) + " GHz" if match else None
```

```python
df['cpu_speed'] = df['info'].apply(extract_processor_speed)
```

storage type **extract**

```python
def extract_storage_type(text):
    match = re.search(r"Storage Type\s*([A-Za-z\s\(\)]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
```

```python
df['storage_type'] = df['info'].apply(extract_storage_type)
```

RAM extract

```python
def extract_ram(text):
    match = re.search(r"RAM Size\s*([\d.]+)\s*GB", text, re.IGNORECASE)
    return match.group(1) if match else None
```

```python
df['ram'] = df['info'].apply(extract_ram)
```

screen size extract

```python
def extract_screen_size(text):
    match = re.search(r"Screen Size\s*([\d.]+)\s*in", text, re.IGNORECASE)
    return match.group(1) if match else None
```

```python
df['screen_size'] = df['info'].apply(extract_screen_size)
```

weight extract

```python
def extract_weight(text):
    match = re.search(r"Weight\s*([\d.]+)\s*(kg|lbs|pounds)", text, re.IGNORECASE)
    return match.group(1) + " " + match.group(2) if match else None
```

```python
df['weight'] = df['info'].apply(extract_weight)
```

SSD and HDD storage extract

```python
def extract_ssd(text):
    match = re.search(r"SSD Capacity\s*([\d.]+)\s*(GB|TB)", text, re.IGNORECASE)
    return match.group(1) + " " + match.group(2) if match else None
def extract_hard_drive(text):
    match = re.search(r"Hard Drive Capacity\s*([\d.]+)\s*(GB|TB)", text, re.IGNORECASE)
    return match.group(1) + " " + match.group(2) if match else None
```

```python
df['storage_hdd'] = df['info'].apply(extract_hard_drive)
df['storage_ssd'] = df['info'].apply(extract_ssd)
```

OS extract

```python
def extract_os(text):
    match = re.search(r"Operating System\s*([A-Za-z0-9\s\-]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
```

```python
df['os'] = df['info'].apply(extract_os)
```

Processor infos, Max resolution, GPU, Model, Year, Color, Type Extract

```python
def extract_max_resolution(text):
    match = re.search(r"Maximum Resolution\s*([\d\sx]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_gpu(text):
    match = re.search(r"GPU\s*([A-Za-z0-9\s\-]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_processor(text):
    match = re.search(r"Processor\s*([A-Za-z0-9\s\-]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_gpu_type(text):
    pattern = r"Graphics Processing Type\s*:\s*([^\n]+)"
    match = re.search(pattern, text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_model(text):
    match = re.search(r"Model\s*([A-Za-z0-9/\- ]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_type(text):
    match = re.search(r"Type\s*([A-Za-z0-9/\- ]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
def extract_year(text):
    match = re.search(r"Release Year\s*(\d{4})", text, re.IGNORECASE)
    return match.group(1) if match else None
def extract_color(text):
    match = re.search(r"Color\s*([A-Za-z\s]+)", text, re.IGNORECASE)
    return match.group(1).strip() if match else None
```

```python
df['type'] = df['info'].apply(extract_type)
df['model'] = df['info'].apply(extract_model)
df['processor'] = df['info'].apply(extract_processor)
df['gpu'] = df['info'].apply(extract_gpu)
df['gpu_type'] = df['info'].apply(extract_gpu_type)
df['max_res'] = df['info'].apply(extract_max_resolution)
df['year'] = df['info'].apply(extract_year)
df['color'] = df['info'].apply(extract_color)
```

Eliminate unused variables

```python
data = df.drop(['title','info','year','model','type','storage','gpu_type','weight','color','max_res'], axis=1)
```
